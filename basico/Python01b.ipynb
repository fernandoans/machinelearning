### Cpias de Listas

```python
lst1 = [0, 2, 4, 6, 8]
lst2 = lst1               # Nova atribuio de nome a varivel
print(lst1, lst2)
lst1.pop(2)
print(lst1, lst2)
```

```python
# Primeira Soluo - Copiar os elementos dessa lista
lst1 = [0, 2, 4, 6, 8]
lst2 = lst1[::] # <---
print(lst1, lst2)
lst1.pop(2)
print(lst1, lst2)
```

```python
# Segunda Soluo - Usar o mtodo deepcopy
from copy import deepcopy
lst1 = [0, 2, 4, 6, 8]
lst2 = deepcopy(lst1) # <---
print(lst1, lst2)
lst1.pop(2)
print(lst1, lst2)
```

## Dicionrios

```python
# CHAVE (no pode ser duplicada) e VALOR
imdb = {
    "Mad Max: Fury Road": 8.1,
    "Spider-Man": 7.3,
    "Ex Machina": 1.7,
    "Venom": 7.1
}
print(imdb)
```

```python
import operator
print(sorted(imdb.items(), key=operator.itemgetter(0)))
print(sorted(imdb.items(), key=operator.itemgetter(1)))
```

```python
# Adicionar
imdb["Deadpool"] = 8.0
print(imdb)
```

```python
# Alterar
imdb["Ex Machina"] = 7.7
print(imdb)
```

```python
# Excluir
del(imdb['Spider-Man'])
print(imdb)
```

```python
# imdb.clear() - Elimina todos os elementos mas no o dicionrio
# print(imdb)
```

```python
print(imdb.keys())
print(imdb.values())
print(imdb)
print(imdb.items())
print(imdb.get("Venom"))
print(imdb["Venom"])
```
