## 1. Linear Regression

A regresso linear  um dos algoritmos de aprendizado de mquina supervisionados do Python que observa recursos contnuos e prev um resultado. Dependendo de sua execuo em uma nica varivel ou em muitos recursos, podemos cham-lo de regresso linear simples ou de regresso linear mltipla.

```python
# Bibliotecas para este modelo
from sklearn import linear_model
```

```python
dados = datasets.load_boston()
print("Dados" , dados.data.shape)
print("Alvo", dados.target.shape)
```

```python
descrever(dados)
# print(dados['DESCR'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(dados.data, dados.target, test_size=0.25, random_state=0)
```

```python
model_linear = linear_model.LinearRegression()
model_linear.fit(X_train, y_train)
```

```python
y_pred_lin = model_linear.predict(X_test)
model_linear.coef_[0]
print('Coeficiente do modelo:', model_linear.coef_)
print('Interceptador do modelo:', model_linear.intercept_)
```

```python
score1 = model_linear.score(X_test, y_test)
score2 = r2_score(y_test, y_pred_lin)
print('Acurcia:', score1, score2)
```

<b>Erro Quadrtico Mdio</b>  til na comparao de estimadores, principalmente se um deles for viciado Se os dois estimadores so no-viciados, o estimador mais eficaz  simplesmente aquele com a menor varincia.

## 4. Support Vector Machines (SVM)

SVM  uma Classificao Supervisionada, um dos algoritmos de aprendizado de mquina mais importantes do Python, que plota uma linha que divide diferentes categorias de seus dados. Neste algoritmo de ML, calculamos o vetor para otimizar a linha. Isso  para garantir que o ponto mais prximo de cada grupo esteja mais distante um do outro. Embora quase sempre achamos que este  um vetor linear, pode ser diferente disso.

```python
# Obter os Dados
dados = datasets.load_iris()
X = dados.data[:, :2]
y = dados.target

# Plotar
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='viridis') # plasma
plt.show()
```

```python
# Reduz a 2 centrides
X = X[:100]
y = y[:100]

# Plotar
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='viridis')
plt.show()
```

```python
from sklearn.svm import SVC # "Support Vector Classifier"
```

```python
model = SVC(kernel='linear', C=1E10)
model.fit(X, y)
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # Fronteiras e Margens
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # Vetores suportados
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')
plot_svc_decision_function(model);
```

## kNN (k-Nearest Neighbors)

Algoritmo para classificao e regresso - principalmente para classificao. De aprendizado supervisionado considera diferentes centrides e usa uma funo geralmente <b>Euclidiana</b> para comparar distncias. Em seguida, analisa os resultados e classifica cada ponto ao grupo para otimiz-lo para colocar com todos os pontos mais prximos. Classifica novos casos usando o voto majoritrio de k de seus vizinhos. O caso que ele atribui a uma classe  o mais comum entre seus K vizinhos mais prximos. Para isso, utiliza uma funo de distncia.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Dados
tempo = ['Sol','Sol','Ambiente','Chuva','Chuva','Chuva','Ambiente','Sol','Sol', 'Chuva','Sol','Ambiente','Ambiente','Chuva']
sensacao = ['Calor','Calor','Calor','Umido','Fresco','Fresco','Fresco','Umido','Fresco','Umido','Umido','Umido','Calor','Fresco']

# Varivel Alvo
y = ['NOK','NOK','OK','OK','OK','NOK','OK','NOK','OK','OK','OK','OK','OK','NOK']
```

```python
# Criar um LabelEncoder
le = preprocessing.LabelEncoder()

# Converter os valores para nmeros
t_encoded = le.fit_transform(tempo)
s_encoded = le.fit_transform(sensacao)
print(t_encoded)
print(s_encoded)
```

```python
# Combinar o tempo e a sensao trmica
X = list(zip(t_encoded, s_encoded))
print(X)
```

```python
# Classificar
model = KNeighborsClassifier(n_neighbors=3)
```

```python
# Treinar o modelo
model.fit(X, y)
```

```python
# Predizer Saida
predicted = model.predict([[0,2]]) # 0:Ambiente, 2:Umido
print(predicted)
```

## k-Means

Algoritmo no supervisionado que resolve o problema de agrupamento. Classifica os dados usando vrios clusters. Os pontos de dados dentro de uma classe so homogneos e heterogneos para grupos de pares.

```python
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:, :4]
y = iris.target
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Comprimento Spala', fontsize=18)
plt.ylabel('Largura Spala', fontsize=18)
plt.show()
```

```python
km = KMeans(n_clusters = 3, random_state=21)
km.fit(X)
```

```python
centers = km.cluster_centers_
print(centers)
```

```python
# This will tell us to which cluster does the data observations belong.
new_labels = km.labels_
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow', edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet',edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
fig.show()
```

## Random Forest

 um conjunto de rvores de Deciso. Para classificar cada novo objeto com base em seus atributos, as rvores votam na classe - cada rvore fornece uma classificao. A classificao com mais votos vence na floresta. 

```python
import pylab as pl
from sklearn.ensemble import RandomForestClassifier
```

```python
x = np.random.uniform(1,100,1000)
y = np.log(x)+np.random.normal(0,.3,1000)
```

```python
pl.scatter(x,y,s=1,label='log(x) with noise')
pl.plot(np.arange(1,100),np.log(np.arange(1,100)),c='b',label='log(x) true function')
pl.xlabel('x')
pl.ylabel('f(x)=log(x)')
pl.legend(loc='best')
pl.title('A basic log function')
pl.show()
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['is_train'] = np.random.uniform(0,1,len(df)) <= .75
df['species'] = pd.Categorical.from_codes(iris.target,iris.target_names)
df.head()
```

```python
train,test = df[df['is_train']==True],df[df['is_train']==False]
features = df.columns[:4]
clf = RandomForestClassifier(n_jobs=2)
y,_ = pd.factorize(train['species'])
clf.fit(train[features],y)
```

```python
preds = iris.target_names[clf.predict(test[features])]
pd.crosstab(test['species'],preds,rownames=['actual'],colnames=['preds'])
```
