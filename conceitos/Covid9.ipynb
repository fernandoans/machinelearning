# COVID-19

```python
# Bibliotecas Necessrias
import pandas as pd
import numpy as np

%matplotlib inline
```

## Dados dos Casos 

```python
casos = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
casos.tail()
```

## Casos no Brasil

```python
casosBrasil = casos[casos['Country/Region'] == 'Brazil']
valCasosBrasil = np.sum(casosBrasil.iloc[ : , 4 : casosBrasil.shape[1]])
valCasosBrasil.plot()
```

## Obiturios

```python
mortes = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
mortes.tail()
```

## Obiturios no Brasil

```python
mortesBrasil = mortes[mortes['Country/Region'] == 'Brazil']
valMortesBrasil = np.sum(mortesBrasil.iloc[ : , 4 : mortesBrasil.shape[1]])
valMortesBrasil.plot()
```

## Clculo da Taxa de Mortalidade no Brasil

```python
mortalidade = (valMortesBrasil/valCasosBrasil)*100
mortalidade.plot()
```

## Pacientes Recuperados

```python
recuperados = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
recuperados.tail()
```

## Pacientes Recuperados no Brasil

```python
recuperadosBrasil = recuperados[recuperados['Country/Region'] == 'Brazil']
valRecuperadosBrasil = np.sum(recuperadosBrasil.iloc[ : , 4 : recuperadosBrasil.shape[1]])
valRecuperadosBrasil.plot()
```

## Comparativo

```python
casosAtivos = valCasosBrasil - (valRecuperadosBrasil + valMortesBrasil)
casosAtivos.plot()
```

```python
df = pd.concat([valCasosBrasil, casosAtivos, valRecuperadosBrasil, valMortesBrasil], axis=1)
df.columns = (["casos", "ativos", "recuperados", "mortes"])
df.plot()
```
