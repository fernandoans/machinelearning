```python
import pandas as pd
```

```python
xlsx = pd.read_excel("bases/ENEM_2017_2019_AMOSTRA_01.xlsx")
xlsx.head()
```

```python
xlsx['NU_ANO'].value_counts()
```

```python
my_tab = pd.crosstab(index=xlsx.NU_ANO, columns="frequency")
my_tab
```

```python
my_tab['cumulative_frequency'] = my_tab.cumsum()
my_tab
```

```python
my_tab['percent'] = my_tab.frequency/my_tab.frequency.sum() * 100
my_tab
```

```python
my_tab['cumulative_percent'] = my_tab.percent.cumsum()
my_tab
```
